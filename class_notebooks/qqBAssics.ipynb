# QQ Plot (Quantile–Quantile Plot)

A QQ plot is a graphical tool used to compare the distribution of a dataset to a theoretical distribution (often the normal distribution) or to another dataset.  
It works by plotting the quantiles of the sample data against the quantiles of the reference distribution

- the data follow the reference distribution, the points will approximately fall on a straight diagonal line
- deviations (curved or S-shaped patterns) indicate departures such as skewness or heavy/light tails  
- Q-Q plots are preferred over histograms because they allow visual detection of subtle distributional differences


```r
sample_data = rnorm(100, mean = 5, sd = 2)

# sort the data
sorted_data = sort(sample_data)
```

```r
# compute the theoretical quantiles of the reference distribution
# eg: a normal distribution with mean = mean(sample), sd = sd(sample)
n = length(sorted_data)
probabilities = (1:n) / (n + 1)   # plotting positions
theoretical_quantiles = qnorm(probabilities, mean = mean(sample_data), sd = sd(sample_data))
```

```r
# plot sample quantiles vs theoretical quantiles
plot(theoretical_quantiles, sorted_data,
     main = "Q-Q from scratch",
     xlab = "Quantiles (Normal)-theory",
     ylab = "Sample Quantiles",
     pch = 19, col = "blue")

# a 45-degree reference line
abline(0, 1, col = "red", lwd = 2)
```

```r
sample_data = rexp(100, rate = 1)  #skewed

#sort the sample data
sorted_data = sort(sample_data)
```

```r
# compute the theoretical quantiles of the normal distribution
n = length(sorted_data)
probabilities = (1:n) / (n + 1)   # plotting positions
theoretical_quantiles = qnorm(probabilities, mean = mean(sample_data), sd = sd(sample_data))
```

```r
# sample quantiles vs normal quantiles
plot(theoretical_quantiles, sorted_data,
     main = "Q-Q Plot: Exponential Data vs Normal",
     xlab = "Theoretical Quantiles (Normal)",
     ylab = "Sample Quantiles (Exponential)",
     pch = 19, col = "blue")

# put the 45-degree reference line
abline(0, 1, col = "red", lwd = 2)
```

```r
# 2 sample case

sample1 = rnorm(100, mean = 5, sd = 2) # normal with mean 5
sample2 = rnorm(80,  mean = 6, sd = 3) # normal with different params

sorted1 = sort(sample1)
sorted2 = sort(sample2)
```

```r
# common probabilities between 0 and 1
n = min(length(sorted1), length(sorted2)) # take smallest length
probabilities = (1:n) / (n + 1)
```

```r
# compute sample quantiles at those probabilities
q1 = quantile(sample1, probs = probabilities, type = 7)
q2 = quantile(sample2, probs = probabilities, type = 7)
```

```r
q2
```

```r
plot(q1, q2,
     main = "two-sample Q-Q Plot (sample1 vs sample2)",
     xlab = "quantiles of sample 1",
     ylab = "quantiles of sample 2",
     pch = 19, col = "blue")

abline(0, 1, col = "red", lwd = 2)
```

```r
x = rnorm(120, mean = 0, sd = 1)   # normal sample
y = rexp(100, rate = 1)            # exponential sample (skewed)
```

```r
# how many quantiles to compare?
n = min(length(x), length(y)) # smallest sample size
```

```r
p = (1:n) / (n + 1) # plotting positions
```

```r
# compute quantiles from both samples at those probabilities
qx = quantile(x, probs = p, type = 7)
qy = quantile(y, probs = p, type = 7)
```

```r
plot(qx, qy,
     main = "two-sample Q-Q: normal vs exponential",
     xlab = "quantiles of normal sample",
     ylab = "quantiles of exponential sample",
     pch = 19, col = "blue")

abline(0, 1, col = "red", lty = 2)
```

```r
x = rnorm(120, mean = 0, sd = 1)  # normal sample (reference on x-axis)
y = rexp(100, rate = 1)     # exponential sample (skewed, on y-axis)

# common probabilities and matched empirical quantiles
n <= min(length(x), length(y))
p = (1:n) / (n + 1)  # plotting positions
qx = as.numeric(quantile(x, probs = p, type = 7))
qy = as.numeric(quantile(y, probs = p, type = 7))

# initialize the plotting device explicitly, then add points
# (type="n" creates the plot window without drawing points yet)
plot(range(qx), range(qy),
     type = "n",
     main = "two-sample QQ: normal vs exponential",
     xlab = "quantiles of normal sample (x)",
     ylab = "quantiles of exponential sample (y)")
points(qx, qy, pch = 19)

# add a reference (equal distributions) line
abline(a = 0, b = 1, lty = 2)

# add a quartile-anchored line (median/IQR-based, robust)
q_x = as.numeric(quantile(x, probs = c(0.25, 0.5, 0.75), type = 7))
q_y = as.numeric(quantile(y, probs = c(0.25, 0.5, 0.75), type = 7))
slope  = (q_y[3] - q_y[1]) / (q_x[3] - q_x[1])   # IQR_y / IQR_x
intercept = q_y[2] - slope * q_x[2]     # med_y - slope * med_x
abline(a = intercept, b = slope, lwd = 2)
```
