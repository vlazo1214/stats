# simple linear regression

## concept

simple linear regression models a linear relationship between one predictor $x$ and a response $y$:

$y_i = \beta_0 + \beta_1 x_i + \varepsilon_i,\quad i=1,\dots,n,$

where $\beta_0$ is the intercept, $\beta_1$ the slope, and $\varepsilon_i$ random errors with mean 0 and variance $\sigma^2$

Ordinary Least Squares (OLS): chooses $\hat\beta_0,\hat\beta_1$ to minimize the residual sum of squares:

$\mathrm{RSS}(\beta_0,\beta_1) = \sum_{i=1}^n\left(y_i - \beta_0 - \beta_1 x_i\right)^2$

closed-form estimates:
$\hat\beta_1 = \frac{\sum_i (x_i-\bar x)(y_i-\bar y)}{\sum_i (x_i-\bar x)^2},\qquad
\hat\beta_0 = \bar y - \hat\beta_1\,\bar x$

fitted values and residuals:

$\hat y_i = \hat\beta_0 + \hat\beta_1 x_i,\qquad e_i = y_i - \hat y_i$

goodness-of-fit:
- **Residual Standard Error (RSE):** $ \widehat\sigma = \sqrt{\frac{1}{n-2}\sum e_i^2} $
- **$R^2$:** fraction of variance explained, $ R^2 = 1 - \frac{\sum e_i^2}{\sum (y_i-\bar y)^2} $

## assumptions

1. **linearity:** $ \mathbb{E}[Y\mid X=x] = \beta_0 + \beta_1 x $.
2. **independence:** observations $(x_i,y_i)$ are independent.
3. **homoscedasticity:** $ \mathrm{Var}(\varepsilon_i) = \sigma^2 $ is constant across $x$.
4. **normality of errors:** $\varepsilon_i \sim \mathcal{N}(0,\sigma^2)$ for valid $t$- and $F$-inference (large samples relax this via CLT).
5. **no high-leverage outliers** that unduly influence the fit.

**look out for:** curvature (nonlinearity), funnel-shaped residuals (heteroscedasticity), influential points (Cook’s distance), omitted variables, extrapolation far beyond observed $x$-range

Interpretation checklist:

    Slope estimate​: direction and magnitude of association

    tt-test on beta estimates​: evidence that slope not equal to 0

    R2: variance explained (context-dependent)

    residual plots: check assumptions; consider transformations or added terms if violated

Common fixes when assumptions fail:

    nonlinearity: add polynomial terms (e.g., x2x2) or transform xx or yy, or use generalized additive models

    heteroscedasticity: use heteroscedasticity-robust standard errors (e.g., sandwich + lmtest) or transform yy

    outliers/influence: investigate and justify inclusion/exclusion; consider robust regression (MASS::rlm)

# multiple linear regression (MLR)

## concept

Multiple linear regression models a response $y$ using $p$ predictors $x_1,\dots,x_p$:
 
$y_i = \beta_0 + \beta_1 x_{i1} + \cdots + \beta_p x_{ip} + \varepsilon_i,\quad i=1,\dots,n,$

with errors $\varepsilon_i$ of mean $0$ and variance $\sigma^2$.

**Matrix form:** Let $\mathbf{y}\in\mathbb{R}^{n}$, $\mathbf{X}\in\mathbb{R}^{n\times (p+1)}$ (first column ones), $\boldsymbol{\beta}\in\mathbb{R}^{p+1}$. 

Then

$ \mathbf{y} = \mathbf{X}\boldsymbol{\beta} + \boldsymbol{\varepsilon},\qquad
\hat{\boldsymbol{\beta}} = (\mathbf{X}^\top \mathbf{X})^{-1}\mathbf{X}^\top \mathbf{y}\ \text{ (when invertible)}$

**interpretation:** $\beta_j$ is the partial effect of $x_j$ on $y$, *holding the other predictors fixed*

## assumptions

1. **Linearity:** $\mathbb{E}[Y\mid \mathbf{X}=\mathbf{x}] = \beta_0 + \sum_{j=1}^p \beta_j x_j$
2. **Independence:** observations are independent
3. **Homoscedasticity:** $\mathrm{Var}(\varepsilon_i) = \sigma^2$ is constant
4. **Normality of errors:** $\varepsilon_i \sim \mathcal{N}(0,\sigma^2)$ for valid $t$/$F$ inference (large $n$ relaxes via CLT)
5. **No perfect multicollinearity:** columns of $\mathbf{X}$ are not exact linear combinations
6. **Well-behaved leverage/influence:** avoid extreme leverage points dominating the fit

## inference and fit quantities

- **Coefficient SEs** from OLS; **$t$-tests** for individual $\beta_j$ (df $= n-p-1$)
- **Overall $F$-test** for $\beta_1=\cdots=\beta_p=0$
- **$R^2$** and **Adjusted $R^2$** (penalizes for number of predictors)
- **Prediction intervals** (preferred for communicating future uncertainty)

## caution

- **Multicollinearity:** unstable coefficients, large SEs; inspect correlations, VIFs
- **Misspecification:** omitted variables, wrong functional form, missing interactions
- **Heteroscedasticity:** funnel patterns in residuals; consider robust SEs or transforms
- **Outliers/Influence:** check leverage, Cook’s distance
- **Extrapolation:** predictions outside the observed predictor ranges are unreliable
- **Data leakage:** perform preprocessing (e.g., scaling/selection) within resampling folds

model building workflow

start from theory and EDA: visualize relationships, check correlations and scatterplots

specify a baseline model: include key confounders; keep it interpretable and parsimonious

refine form: add interactions and nonlinear terms only with substantive or diagnostic justification

validate: use cross-validation or a holdout 

report diagnostics and uncertainty: residual checks, influence measures, VIFs, CIs, prediction intervals

when OLS is Not Enough

regularization: ridge, lasso, elastic net (eg, via glmnet) for high-
p or collinearity

generalized linear models: logistic, Poisson, etc., for non-Gaussian responses

robust regression: MASS::rlm to mitigate outliers/heavy tails

nonparametric/semiparametric: GAMs (mgcv) for smooth nonlinear effects

find a good regression

````r
n = 300
````

````r
x_strong = runif(n, 0, 10)


x_correlated = 0.6 * scale(x_strong) + 0.4 * rnorm(n)
x_correlated = as.numeric(scale(x_correlated))  
x_correlated = 5 + 2 * x_correlated 
````

````r
x_weak = rnorm(n, mean = 0, sd = 1)
````

````r
x_quad = runif(n, -3, 3)
````

````r
x_loglike = rexp(n, rate = 0.5)  # skewed, positive
x_loglike = pmin(x_loglike, 30)  # cap extreme tail
````

````r
x_noise = rnorm(n)
````

````r
x_group = factor(sample(c("A", "B", "C"), 
                    size = n, replace = TRUE, 
                    prob = c(0.4, 0.35, 0.25)))
````

````r
beta0 = 3.0
eps = rnorm(n, sd = 1.5)
````

````r
y = beta0 + 1.2 * x_strong + 0.8 * x_correlated + 1.5 * (x_quad^2) +
     2.0 * log1p(x_loglike) + ifelse(x_group == "B",  1.0, 0) +
     ifelse(x_group == "C", -0.7, 0) +
     0.1 * x_weak +          # tiny linear effect
     0.0 * x_noise +         # zero true effect
     eps
````

````r
df = data.frame(
  y = y,
  x_strong = x_strong,
  x_correlated = x_correlated,
  x_weak = x_weak,
  x_quad = x_quad,
  x_loglike = x_loglike,
  x_noise = x_noise,
  x_group = x_group
)
````

````r
missing_indices = sample(seq_len(n), size = 8)
df$x_weak[missing_indices[1:3]] = NA
df$x_loglike[missing_indices[4:6]] = NA
df$y[missing_indices[7:8]] = NA
````

````r
fit_results = list()

fit_one_slr = function(formula_str, data) {
  fit = lm(as.formula(formula_str), data = data)
  s   = summary(fit)
  # slope p-value if numeric predictor, for factor-only, use overall F-test p-value
  p_value = NA_real_
  slope_name = setdiff(names(coef(fit)), "(Intercept)")
  if (length(slope_name) == 1 && slope_name %in% rownames(s$coefficients)) {
    p_value = s$coefficients[slope_name, "Pr(>|t|)"]
  } else {
    #
    if (!is.null(s$fstatistic)) {
      p_value = pf(s$fstatistic[1], s$fstatistic[2], s$fstatistic[3], lower.tail = FALSE)
    }
  }
  list(
    fit = fit,
    formula = formula_str,
    r2 = s$r.squared,
    adj_r2 = s$adj.r.squared,
    sigma = s$sigma,
    p_value = p_value
  )
}
````

````r
candidates = c(
  "y ~ x_strong",
  "y ~ x_correlated",
  "y ~ x_weak",
  "y ~ x_quad",
  "y ~ x_loglike",
  "y ~ x_noise",
  "y ~ x_group"     
)

for (f in candidates) {
  fit_results[[f]] = fit_one_slr(f, df)
}
````

````r
# rank models by R^2 (descending)
ranked = do.call(rbind, lapply(fit_results, function(res) {
  cbind.data.frame(
    formula = res$formula,
    r2 = res$r2,
    adj_r2 = res$adj_r2,
    sigma = res$sigma,
    p_value = res$p_value,
    stringsAsFactors = FALSE
  )
}))

ranked = ranked[order(-ranked$r2), ]
print(ranked, row.names = FALSE)
````

````r
plot_top_k = 3
numeric_models = ranked[!grepl("x_group", ranked$formula), , drop = FALSE]
top_k = head(numeric_models, plot_top_k)

plot_slr_with_bands = function(formula_str, data) {
  fit = lm(as.formula(formula_str), data = data)
  
  parts = strsplit(formula_str, "~")[[1]]
  predictor = trimws(parts[2])

  #
  dd = data[complete.cases(data[, c("y", predictor)]), c("y", predictor)]
  names(dd) = c("y", "x")

  # grid and intervals
  x_grid = seq(min(dd$x), max(dd$x), length.out = 200)
  pred_conf = predict(fit, newdata = setNames(data.frame(x_grid), predictor),
                       interval = "confidence", level = 0.95)
  pred_pred = predict(fit, newdata = setNames(data.frame(x_grid), predictor),
                       interval = "prediction", level = 0.95)

  #
  plot(dd$x, dd$y, pch = 19, cex = 0.8,
       xlab = predictor, ylab = "y",
       main = paste0("SLR: ", formula_str))
  polygon(c(x_grid, rev(x_grid)),
          c(pred_pred[,"lwr"], rev(pred_pred[,"upr"])),
          border = NA, col = rgb(0, 0, 0, 0.09))
  polygon(c(x_grid, rev(x_grid)),
          c(pred_conf[,"lwr"], rev(pred_conf[,"upr"])),
          border = NA, col = rgb(0, 0, 0, 0.18))
  lines(x_grid, pred_conf[,"fit"], lwd = 2)
  legend("topleft",
         legend = c("Observed", "Fitted", "95% CI (mean)", "95% PI (new)"),
         pch = c(19, NA, 15, 15), lty = c(NA, 1, NA, NA), lwd = c(NA, 2, NA, NA),
         col = c("black", "black", rgb(0,0,0,0.18), rgb(0,0,0,0.09)), bty = "n")
}

if (nrow(top_k) > 0) {
  par(mfrow = c(min(2, nrow(top_k)), 1), mar = c(4,4,2,1))
  apply(top_k, 1, function(row) plot_slr_with_bands(row["formula"], df))
  par(mfrow = c(1,1))
}
````

````r
plot_group_effect = function(y, g, main_title = "y by x_group (means ± SE)") {
  ok = complete.cases(y, g)
  y = y[ok]; g = droplevels(g[ok])

  group_means = tapply(y, g, mean)
  group_sds   = tapply(y, g, sd)
  group_ns    = tapply(y, g, length)
  group_se    = group_sds / sqrt(group_ns)

  bp = barplot(group_means, ylim = c(min(group_means - 2*group_se), max(group_means + 2*group_se)),
                ylab = "Mean of y", main = main_title, xlab = "Group")
  arrows(bp, group_means - group_se, bp, group_means + group_se, angle = 90, code = 3, length = 0.06)
}

plot_group_effect(df$y, df$x_group)
````

# Multicollinearity: Concepts, Tests, Interpretation, and Remedies

Multicollinearity occurs in **multiple linear regression** when two or more predictors convey overlapping information (are linearly dependent or nearly so). It does **not** affect the *unbiasedness* of OLS coefficients, but it **inflates their variance**, making estimates unstable and hypothesis tests unreliable

---

## why it matters

- **Unstable coefficients:** small data changes cause large swings in estimates
- **Large standard errors:** individually “insignificant” predictors despite strong joint effect
- **Counterintuitive signs:** coefficients may flip sign even when the marginal correlation suggests otherwise
- **Wide confidence intervals and noisy predictions** when extrapolating in directions poorly supported by the data

---

## when it applies

- **Multiple regression only.** With a single predictor, multicollinearity is not defined
- Especially common when:
  - Predictors are constructed from each other (e.g., `x` and `x + small_noise`)
  - One predictor is almost a linear combination of others (polynomial terms without centering; interactions without centering)
  - High-dimensional settings (many predictors vs. sample size)
  - Features that measure similar constructs (e.g., height in cm and inches)

---

## quick diagnostics (from simplest to more formal)

1. **pairwise correlations**  
   - large absolute correlations (e.g., $|r| > 0.8$) among predictors are a *signal*
   - limitation: multicollinearity can exist even if all pairwise correlations are modest (e.g., with 3+ predictors)

2. **tolerance and VIF (Variance Inflation Factor)**  
   - For predictor $X_j$, regress $X_j$ on all other predictors; call the $R_j^2$ from that regression
   - **tolerance:** $ \text{Tol}_j = 1 - R_j^2 $. Small tolerance indicates redundancy
   - **VIF:** $ \text{VIF}_j = \frac{1}{1 - R_j^2} = \frac{1}{\text{Tol}_j}. $
   - Interpretation (rules of thumb, context matters):
     - VIF ~ 1–2: low concern  
     - VIF ~ 5: moderate concern (investigate)  
     - VIF >= 10: high concern (often problematic)

3. **Condition number (κ) of the design matrix**  
   - Compute the condition number of the **scaled** design matrix $X$ (or of $X^\top X$).  
   - κ measures how close $X$ is to being rank-deficient.
   - Interpretation (rules of thumb):
     - κ < 10–15: likely fine  
     - κ ≈ 15–30: moderate concern  
     - κ > 30–50: serious multicollinearity  
   - **Scale matters:** standardize predictors before interpreting κ.

4. **Eigen-analysis / variance-decomposition proportions (advanced)**  
   - Examine small eigenvalues of $X^\top X$ and how coefficient variance is distributed across eigen-components.  
   - Useful for diagnosing which subsets of predictors co-move.

---

## Minimal R examples (base R)

> These snippets illustrate how to compute pairwise correlations, a **crude VIF** without extra packages, and a **condition number**. Replace `df` and predictor names as needed

### 1) Pairwise correlations among predictors
```r
predictor_names = c("x1", "x2", "x3", "x4")
cor_mat = cor(df[, predictor_names], use = "pairwise.complete.obs")
round(cor_mat, 3)
```

### 2) Crude VIFs

```r
# Fit your regression first
fit = lm(y ~ x1 + x2 + x3 + x4, data = df)

# Model matrix (includes intercept as the first column)
X = model.matrix(fit)
pred_cols = colnames(X)[-1]  # drop intercept

vif_values = numeric(length(pred_cols))
for (j in seq_along(pred_cols)) {
  target = pred_cols[j]
  others = pred_cols[-j]
  dd = as.data.frame(X[, c(target, others), drop = FALSE])
  form = as.formula(paste(target, "~", paste(others, collapse = " + ")))
  r2 = summary(lm(form, data = dd))$r.squared
  vif_values[j] = 1 / (1 - r2)
}
data.frame(predictor = pred_cols, VIF = round(vif_values, 2))
```

### 3) Condition number k (scaled)

```r
# center/scale numeric predictors before computing kappa
Z = scale(df[, predictor_names])
# add intercept column for consistency (optional for k)
X_scaled = cbind(Intercept = 1, Z)
kappa_value = kappa(X_scaled)   # or kappa(cor(Z)) for correlation-based k
kappa_value
```

Interpreting outputs:

- VIFs: look for unusually large values (eg, >= 5-10) focus on the subset of predictors with the largest VIF
- k: if k is well above ~30-50 (after scaling), multicollinearity is likely severe


Interpretation checklist

- Are coefficients unstable or counterintuitive? Compare signs and magnitudes to subject-matter expectations
- Are standard errors large for some predictors despite strong overall model fit?
This pattern often contains multicollinearity
- Do VIFs or k flag an issue? Use them as screening tools, not concrete rules


solutions

- Combine redundant features or pick one representative
- Centering helps with polynomials and interactions (reduces non-essential collinearity among x,x2,x times z,x ,x2,x times z); it does not reduce exact linear dependence
- orthogonal polynomials for polynomial regression (poly(x, degree, raw = FALSE) in R)
- ridge regression (L2) stabilizes coefficients under multicollinearity by shrinking them
- Lasso / Elastic Net can both regularize and perform variable selection
- Principal components or partial least squares
- Re-express predictors in low-dimensional orthogonal components; interpretability shifts from original variables to components


Multicollinearity is about predictor redundancy; it inflates variance of coefficient estimates.

# Using ANOVA for Model Selection (Nested Models)

**Goal:** Use ANOVA to compare **nested** regression models as a principled way to decide whether the added terms (predictors, interactions, or higher-order polynomials) provide a statistically meaningful improvement in fit

---

## What ANOVA Does in This Context

For **Gaussian linear models** (`lm`), `anova(model_small, model_large)` performs a **partial F-test** of:

$
H_0:\ \text{all extra coefficients in the larger model} = 0
\quad\text{vs}\quad
H_1:\ \text{at least one extra coefficient} \neq 0.
$

- **Nested models** means: every term in `model_small` is also in `model_large`
- the test compares **residual sums of squares (RSS)** and **degrees of freedom**; the resulting **F-statistic** and **p-value** tell you if the larger model significantly improves fit
- ff **p-value < α** (commonly 0.05), prefer the **larger** model; otherwise, prefer the **smaller** model for parsimony

> for **GLMs** (e.g., logistic regression), use `anova(..., test = "Chisq")`, which performs a **likelihood-ratio test** based on deviances

---

## Workflow Pattern

1. **Start with a baseline model** guided by subject-matter knowledge
2. **Add one or more terms** (eg, a suspected interaction or curvature via polynomial terms)
3. **Compare with ANOVA** to test if the extra terms improve fit
4. **Repeat incrementally** (hierarchical principle: include lower-order terms when adding interactions or polynomials)
5. **Balance statistics and parsimony**; corroborate with diagnostics and out-of-sample checks (e.g., CV)

---

### Example: Gaussian Linear Models (F-tests)

```r
n = 150
x1 = runif(n, 0, 10)
x2 = rnorm(n)
y  = 2 + 1.2*x1 + 0.5*x2 + 0.6*x1*x2 + rnorm(n, 0, 1.5)
df = data.frame(y = y, x1 = x1, x2 = x2)

# (no interaction)
m0 = lm(y ~ x1 + x2, data = df)

# adds interaction
m1 = lm(y ~ x1 * x2, data = df)  # equivalent to y ~ x1 + x2 + x1:x2

# ANOVA comparison (nested F-test)
anova(m0, m1)  # if p-value is small, keep the interaction (m1)
summary(m1)    # inspect coefficients and SEs if interaction is retained
```

Interpretation:

if anova(m0, m1) yields a small p-value, the interaction term improves model fit; prefer m1

if not significant, prefer the simpler m0 unless theory dictates otherwise

```r
x  = runif(n, -3, 3)
y  = 1 + 0.8*x + 1.5*x^2 + rnorm(n, 0, 1.2)
df2 = data.frame(y = y, x = x)

# linear-only model
l0 = lm(y ~ x, data = df2)

# add quadratic term (hierarchical: keep x when adding x^2)
l1 = lm(y ~ x + I(x^2), data = df2)

anova(l0, l1)  # significant? If yes, retain curvature (l1)
```

always keep lower-order terms when adding higher-order terms (x^2, x^3) or interactions (x1:x2)

Example

```r
# start simple, then add blocks of terms stepwise (hierarchical)
m_base = lm(y ~ x1 + x2, data = df)
m_int  = lm(y ~ x1 * x2, data = df)           # add interaction
m_poly = lm(y ~ poly(x1, 2, raw = TRUE) + x2, data = df)  # add curvature on x1

# pairwise nested comparisons (only compare nested pairs)
anova(m_base, m_int)
anova(m_base, m_poly)
# m_int and m_poly are not nested in each other; do not compare them via anova()
```

(important: ANOVA only compares nested models. for non-nested models, use information criteria (AIC/BIC) or cross-validation)

### Dropping Terms Within a Model (Partial F-tests)

use drop1 to test the incremental contribution of each term in a chosen model:

```r
full = lm(y ~ x1 * x2, data = df)
drop1(full, test = "F")  # partial F-test for each term given the others
```

interpretation:

a small p-value for a term (e.g., x1:x2) indicates it contributes beyond the other terms already in the model
