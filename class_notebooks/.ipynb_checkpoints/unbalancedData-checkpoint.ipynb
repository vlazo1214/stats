```r
#install.packages("ROSE")
library(ROSE)
```

```r
n = 1000
X1 = rnorm(n)
X2 = rnorm(n)

threshold = quantile(X1 + X2 + rnorm(n), 0.98)
Y = ifelse(X1 + X2 + rnorm(n) > threshold, 1, 0)

data = data.frame(X1, X2, Y)

# Display class distribution
table(data$Y)
```

```r
?ovun.sample
```

```r
data_balanced = ovun.sample(Y ~ ., data = data, p=0.5 , method = "over")$data  # Adjust N as needed
table(data_balanced$Y)
```

```r
indices = sample(1:nrow(data_balanced), nrow(data_balanced)*0.7)
train_data = data_balanced[indices, ]
test_data = data_balanced[-indices, ]
```

```r
model = glm(Y ~ ., data=train_data, family=binomial)
summary(model)
```

```r
probabilities = predict(model, newdata=test_data, type="response")
predictions = ifelse(probabilities > 0.5, 1, 0)
```

```r
library(caret)
confusionMatrix( as.factor(test_data$Y), as.factor(predictions) )
```

```r
cm = confusionMatrix(as.factor(predictions), as.factor(test_data$Y))
#compute the F1 score
precision = cm$byClass["Pos Pred Value"]
recall = cm$byClass["Sensitivity"]
f1_score = 2 * (precision * recall) / (precision + recall)
print(f1_score)
```

```r
library(pROC)
```

```r
roc_obj = roc(test_data$Y, probabilities)
```

```r
plot(roc_obj, main="ROC Curve", col="blue", lwd=2)
abline(h=0, v=1, col="gray", lty=2)
```

```r
auc(roc_obj)
```

```r
indices = sample(1:nrow(data), nrow(data)*0.7)
train_data_unbalanced = data[indices, ]
test_data_unbalanced = data[-indices, ]

model_unbalanced = glm(Y ~ ., data=train_data_unbalanced, family=binomial)
```

```r
probabilities_unbalanced = predict(model_unbalanced, newdata=test_data_unbalanced, type="response")
predictions_unbalanced = ifelse(probabilities_unbalanced > 0.5, 1, 0)
```

```r
cm_unbalanced = confusionMatrix(as.factor(predictions_unbalanced), as.factor(test_data_unbalanced$Y))
accuracy_unbalanced = cm_unbalanced$overall["Accuracy"]
print(accuracy_unbalanced)
```

```r
roc_obj_unbalanced = roc(test_data_unbalanced$Y, probabilities_unbalanced)
```

```r
plot(roc_obj_unbalanced, main="ROC Curve (Unbalanced Data)", col="red", lwd=2)
abline(h=0, v=1, col="gray", lty=2)
```

```r
auc(roc_obj_unbalanced)
```

```r
cm_unbalanced <- confusionMatrix(as.factor(predictions_unbalanced), as.factor(test_data_unbalanced$Y))
precision_unbalanced <- cm_unbalanced$byClass["Pos Pred Value"]
recall_unbalanced <- cm_unbalanced$byClass["Sensitivity"]
f1_score_unbalanced <- 2 * (precision_unbalanced * recall_unbalanced) / (precision_unbalanced + recall_unbalanced)
print(f1_score_unbalanced)
```
