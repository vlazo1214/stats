```r
n      = 20       # sample size
mu0    = 0        # null mean
mu_data= 0.4      # true mean (alternative)
sd_pop = 1
alpha  = 0.1
nsim   = 5000     # number of simulated experiments
```

```r
t_null <- numeric(nsim)
for (i in 1:nsim) {
  x = rnorm(n, mean = mu0, sd = sd_pop)
  t_null[i] = (mean(x) - mu0) / (sd(x) / sqrt(n))
}
```

```r
crit_lo = quantile(t_null, alpha/2)
crit_hi = quantile(t_null, 1 - alpha/2)
```

```r
crit_hi
```

```r
# simulate alternative distribution of t-statistics
t_alt = numeric(nsim)
for (i in 1:nsim) {
  x = rnorm(n, mean = mu_data, sd = sd_pop)
  t_alt[i] = (mean(x) - mu0) / (sd(x) / sqrt(n))
}
```

```r
x_obs = rnorm(n, mean = mu_data, sd = sd_pop)
t_obs = (mean(x_obs) - mu0) / (sd(x_obs) / sqrt(n))
```

```r
hist(t_null, breaks = 50, freq = FALSE, col = rgb(1,0,0,0.3),
     main = "Simulated distributions of t-statistic",
     xlab = "t value", ylim = c(0, 0.4))
hist(t_alt, breaks = 50, freq = FALSE, col = rgb(0,0,1,0.3), add = TRUE)
abline(v = c(crit_lo, crit_hi), col = "red", lwd = 2, lty = 2)
abline(v = t_obs, col = "black", lwd = 4)
legend("topright",
       legend = c("Null (H0)", "Alternative (True)", "Critical cutoffs", "Observed t"),
       col = c("red","blue","red","black"),
       lwd = c(10,10,2,2), lty = c(1,1,2,1), bty = "n")
```

```r

xrange = seq(-2, 2, length.out = 200)
plot(xrange, dnorm(xrange, mu0, sd_pop/sqrt(n)), type = "l", lwd = 2, col = "red",
     main = "Sampling distributions of the mean",
     ylab = "Density", xlab = "Sample mean")
lines(xrange, dnorm(xrange, mu_data, sd_pop/sqrt(n)), lwd = 2, col = "blue")
abline(v = mu0, col = "red", lty = 2)
abline(v = mu_data, col = "blue", lty = 2)
legend("topright", legend = c("Null (μ0)", "Alternative (μ_true)"),
       col = c("red","blue"), lwd = 2, bty = "n")

cat("Observed t =", round(t_obs, 3), "\n")
cat("Critical cutoffs (simulated) =", round(crit_lo, 3), "and", round(crit_hi, 3), "\n")
if (t_obs < crit_lo | t_obs > crit_hi) {
  cat("Decision: Reject H0\n")
} else {
  cat("Decision: Fail to reject H0\n")
}
```

```r
# TWO SAMPLE CASE

n1     = 20         # sample size group 1
n2     = 20         # sample size group 2
mu1_H0 = 0          # null mean group 1
mu2_H0 = 0          # null mean group 2 (same under H0)
mu1_H1 = 0          # true mean group 1
mu2_H1 = 0.5        # true mean group 2 (shift under alternative)
sd_pop = 1          # population SD (same for both groups)
alpha  = 0.05
nsim   = 5000       # number of simulations
```

```r
two_sample_t <- function(x, y) {
  n1 = length(x); n2 <- length(y)
  s1 = var(x); s2 <- var(y)
  sp2 = ((n1 - 1) * s1 + (n2 - 1) * s2) / (n1 + n2 - 2)  # pooled variance
  (mean(x) - mean(y)) / sqrt(sp2 * (1/n1 + 1/n2))
}
```

```r
t_null = numeric(nsim)

for (i in 1:nsim) {
  x = rnorm(n1, mean = mu1_H0, sd = sd_pop)
  y = rnorm(n2, mean = mu2_H0, sd = sd_pop)
  t_null[i] = two_sample_t(x, y)
}
```

```r
df = n1 + n2 - 2
crit_lo = quantile(t_null, alpha/2)
crit_hi = quantile(t_null, 1 - alpha/2)
```

```r
t_alt = numeric(nsim)

for (i in 1:nsim) {
  x = rnorm(n1, mean = mu1_H1, sd = sd_pop)
  y = rnorm(n2, mean = mu2_H1, sd = sd_pop)
  t_alt[i] = two_sample_t(x, y)
}
```

```r
x_obs = rnorm(n1, mean = mu1_H1, sd = sd_pop)
y_obs = rnorm(n2, mean = mu2_H1, sd = sd_pop)
t_obs = two_sample_t(x_obs, y_obs)
```

```r
# null vs Alternative distributions of t statistic
hist(t_null, breaks = 50, freq = FALSE, col = rgb(1,0,0,0.3),
     main = "Simulated null vs alternative (two-sample t)",
     xlab = "t statistic", ylim = c(0, 0.4))
hist(t_alt, breaks = 50, freq = FALSE, col = rgb(0,0,1,0.3), add = TRUE)
abline(v = c(crit_lo, crit_hi), col = "red", lwd = 2, lty = 2)
abline(v = t_obs, col = "black", lwd = 2)
legend("topright",
       legend = c("Null (H0)", "Alternative (H1)", "Critical cutoffs", "Observed t"),
       col = c("red","blue","red","black"),
       lwd = c(10,10,2,2), lty = c(1,1,2,1), bty = "n")
```

```r

# population distributions under H0 vs H1
xrange = seq(-2, 2, length.out = 300)
plot(xrange, dnorm(xrange, mean = mu1_H0, sd = sd_pop), type = "l",
     col = "red", lwd = 2,
     main = "Population distributions of Group 1 and Group 2",
     ylab = "Density", xlab = "Value", ylim = c(0, 0.45))
lines(xrange, dnorm(xrange, mean = mu2_H0, sd = sd_pop), col = "red", lwd = 2, lty = 2)
lines(xrange, dnorm(xrange, mean = mu1_H1, sd = sd_pop), col = "blue", lwd = 2)
lines(xrange, dnorm(xrange, mean = mu2_H1, sd = sd_pop), col = "blue", lwd = 2, lty = 2)
legend("topright",
       legend = c("Group 1", "Group 2"),
       col = c("blue","blue"), lwd = 2, lty = c(1,2), bty = "n")
```

```r
n        = 30        # sample size
alpha    = 0.05      # significance level
nsim     = 5000      # number of simulations for null/alt
beta0    = 0.0       # intercept
beta1_H0 = 0.0       # null slope   (H0: beta1 = 0)
beta1_H1 = 0.65       # alt  slope   (H1: beta1 != 0)
sigma    = 1.0       # error SD
```

```r
x = seq(-1, 1, length.out = n)
```

```r
f_statistic = function(x, y) {
  n  = length(y)
  xbar = mean(x); 
  ybar = mean(y)
  Sxx = sum( (x - xbar)^2 )
  Sxy = sum( (x - xbar) * (y - ybar) )
  beta1_hat = Sxy / Sxx
  beta0_hat = ybar - beta1_hat * xbar

  yhat <- beta0_hat + beta1_hat * x
  
  # ANOVA decomposition
  SSR = sum( (yhat - ybar)^2 )         # regression sum of squares
  SSE = sum( (y - yhat)^2 )            # residual sum of squares
  df1 = 1
  df2 = n - 2
  F   = (SSR / df1) / (SSE / df2)
  c(F = F, df1 = df1, df2 = df2, b0 = beta0_hat, b1 = beta1_hat, ybar = ybar)
}
```

```r
F_null = numeric(nsim)
for (i in 1:nsim) {
  y0 = beta0 + rnorm(n, 0, sigma)
  F_null[i] = f_statistic(x, y0)[["F"]]
}
```

```r
crit_hi = as.numeric(quantile(F_null, 1 - alpha))
crit_hi
```

```r
F_alt = numeric(nsim)

for (i in 1:nsim) {
  y1 = beta0 + beta1_H1 * x + rnorm(n, 0, sigma)
  F_alt[i] = f_statistic(x, y1)[["F"]]
}
```

```r
y_obs = beta0 + beta1_H1 * x + rnorm(n, 0, sigma)
stats = f_statistic(x, y_obs)
F_obs = stats[["F"]]; 
df1 = stats[["df1"]]; 
df2 = stats[["df2"]]
```

```r
hist(F_null, breaks = 50, freq = FALSE, col = rgb(1,0,0,0.3),
     main = "Simulated F-statistic: Null vs Alternative",
     xlab = "F statistic")
hist(F_alt, breaks = 50, freq = FALSE, col = rgb(0,0,1,0.3), add = TRUE)
abline(v = crit_hi, col = "red", lwd = 2, lty = 2)
abline(v = F_obs,   col = "black", lwd = 2)
legend("topright",
       legend = c("Null (H0: slope=0)", "Alternative (H1: slope≠0)", 
                  "Critical cutoff (empirical)", "Observed F"),
       col = c("red","blue","red","black"),
       lty = c(1,1,2,1), lwd = c(10,10,2,2), bty = "n")
```

```r
plot(x, y_obs, pch = 19, col = "grey30",
     xlab = "x", ylab = "y",
     main = "Observed data: fitted line vs null line")
abline(a = stats[["b0"]], b = stats[["b1"]], col = "blue", lwd = 2)      # fitted line
abline(h = mean(y_obs), col = "red", lwd = 2, lty = 2)                   # null model (slope=0)
```

```r
n       = 100        # sample size
k       = 4          # number of categories
alpha   = 0.05
nsim    = 5000       # number of simulations
```

```r
p_null = rep(1/k, k)
p_null
```

```r
p_alt = c(0.4, 0.3, 0.2, 0.1)  # must sum to 1
```

```r
chi_sq_stat = function(counts, probs, n) {
  expected = n * probs
  sum((counts - expected)^2 / expected)
}
```

```r
chi_null = numeric(nsim)

for (i in 1:nsim) {
  counts = table(factor(sample(1:k, n, replace = TRUE, prob = p_null), levels = 1:k))
  chi_null[i] = chi_sq_stat(as.numeric(counts), p_null, n)
}
```

```r
hist(chi_null)
```

```r
df = k - 1
crit_hi = quantile(chi_null, 1 - alpha)   # empirical cutoff
crit_hi
```

```r
chi_alt = numeric(nsim)

for (i in 1:nsim) {
  counts = table(factor(sample(1:k, n, replace = TRUE, prob = p_alt), levels = 1:k))
  chi_alt[i] = chi_sq_stat(as.numeric(counts), p_null, n)
}
```

```r
counts_obs = table(factor(sample(1:k, n, replace = TRUE, prob = p_alt), levels = 1:k))
chi_obs = chi_sq_stat(as.numeric(counts_obs), p_null, n)
counts_obs
chi_obs
```

```r
hist(chi_null, breaks = 50, freq = FALSE, col = rgb(1,0,0,0.3),
     main = "Simulated Chi-square: Null vs Alternative",
     xlab = "Chi-square statistic")
hist(chi_alt, breaks = 50, freq = FALSE, col = rgb(0,0,1,0.3), add = TRUE)
abline(v = crit_hi, col = "red", lwd = 2, lty = 2)
abline(v = chi_obs, col = "black", lwd = 2)
legend("topright",
       legend = c("Null (H0)", "Alternative (H1)", "Critical cutoff", "Observed Chi-square"),
       col = c("red","blue","red","black"),
       lwd = c(10,10,2,2), lty = c(1,1,2,1), bty = "n")
```

```r
barplot(rbind(Observed = counts_obs, Expected = n * p_null),
        beside = TRUE, col = c("grey60","white"),
        main = "Observed vs Expected Counts",
        ylab = "Counts", names.arg = paste("Cat", 1:k))
legend("topright", legend = c("Observed","Expected"), fill = c("grey60","white"), bty = "n")
```
