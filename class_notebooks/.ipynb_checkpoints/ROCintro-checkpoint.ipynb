# ROC curves - Receiver operating characteristic

help explore the quality of a binary classifier based upon the threshold for the class determination. At the base of it is the classifier task between predicting the positive (P) and negative conditions (N). A True Positive (TP) is when the classifier correctly predicts a Positive instance, and False Positive (FP) when it predicts positive when it should have predicted a negative. The True Negative (TN) is when the classifier correctly predicted a negative instance, and False Negative (FN) when it should have predicted a positive. 

## TPR = True Positive Rate (aka, sensitivity/recal/probability of detection)

this is on the y axis and is found via:

$ TPR = \frac{TP}{P} = \frac{TP}{TP+FN} $

## TNR = True Negative Rate (aka, specificity/selectivity/false alarm probability)

this is on the x axis and is found via: 

$ TNR = \frac{TN}{N} = \frac{TN}{TN+FP} $

The accuracy is $\frac{TP+TN}{P+N}$

```r
n = 1000
x = rnorm(n)
z = 1 + 2*x
prob = 1 / (1 + exp(-z)) #sigmoid
y = rbinom(n, 1, prob)
```

```r
head( y, 20 )
```

```r
model = glm(y ~ x, family = binomial())
summary(model)
```

```r
predicted_probs = predict(model, type = "response")
head( predicted_probs )
```

```r
tpr = numeric()
fpr = numeric()
```

```r
for(threshold in seq(0, 1, by=0.05)) {
  predicted_class = ifelse(predicted_probs > threshold, 1, 0)
  
  TP = sum(predicted_class == 1 & y == 1)
  FP = sum(predicted_class == 1 & y == 0)
  TN = sum(predicted_class == 0 & y == 0)
  FN = sum(predicted_class == 0 & y == 1)
  
  tpr = c(tpr, TP / (TP + FN))
  fpr = c(fpr, FP / (FP + TN))
}
```

```r
plot(fpr, tpr, type = "l", col = "blue", 
     xlab = "False Positive Rate", ylab = "True Positive Rate", 
     main = "ROC Curve")

abline(0, 1, col = "red", lty = 2)
```

```r
generate_table = function(predictions, ground_truth, threshold) {
  predicted_class = ifelse(predictions > threshold, 1, 0)
  table(Actual = ground_truth, Predicted = predicted_class)
}

#thresholds 0.2, 0.5, and 0.8
thresholds = c(0.2, 0.5, 0.8)
for(threshold in thresholds) {
  cat(paste("Table for Threshold:", threshold), "\n")
  print(generate_table(predicted_probs, y, threshold))
  cat("\n")
}
```

```r
generate_confusion_metrics = function(predictions, ground_truth, threshold) {
  predicted_class = ifelse(predictions > threshold, 1, 0)
  cm = table(Actual = ground_truth, Predicted = predicted_class)
  
  # values from table related to the FP etc
  TN = cm[1,1]
  FP = cm[1,2]
  FN = cm[2,1]
  TP = cm[2,2]
  
  #metrics
  accuracy = (TP + TN) / (TP + TN + FP + FN)
  
  list(
    ConfusionMatrix = cm,
    FalsePositives = FP,
    FalseNegatives = FN,
    Accuracy = accuracy
  )
}
```

```r
thresholds = c(0.2, 0.5, 0.8)
for(threshold in thresholds) {
  cat(paste("Metrics for Threshold:", threshold), "\n")
  metrics = generate_confusion_metrics(predicted_probs, y, threshold)
  print(metrics$ConfusionMatrix)
  cat("False Positives:", metrics$FalsePositives, "\n")
  cat("False Negatives:", metrics$FalseNegatives, "\n")
  cat("Accuracy:", round(metrics$Accuracy, 4), "\n")
  cat("\n------------------------\n")
}
```

```r
#compute the AUC manually, values between 0.5 and 1 (best)
sorted_indices = order(fpr)
fpr_sorted = fpr[sorted_indices]
tpr_sorted = tpr[sorted_indices]
auc_manual = sum(diff(fpr_sorted) * (tpr_sorted[-1] + tpr_sorted[-length(tpr_sorted)]) / 2)
```

```r
print(auc_manual)
```

```r
#using  package pROC
roc_obj = roc(y, predicted_probs)

#get the AUC
auc_value = auc(roc_obj)
print(auc_value)
```

```r
plot(roc_obj, main = paste("ROC and AUC =", round(auc_value, 4)))
```

```r
data = data.frame(x=x, y=as.factor(y))
```

```r
rf_model = randomForest(y~x, data=data, ntree=500)
```

```r
rf_probs = predict(rf_model, type="prob")[,2]
```

```r
roc_rf = roc(y, rf_probs)
auc_rf = auc(roc_rf)
```

```r
plot(roc_rf, main=paste("Random Forest ROC Curve - AUC:", round(auc_rf, 4)))
```

## recall

proportion of actual positive cases that were correctly identified by the model

$Recall = \frac{TP}{TP + FN}$

and the 

## F1 Score

harmonic mean of precision and recall. It provides a balance between the two, especially when the classes are unevenly distributed

$F1score = 2 \times \frac{precision \times recall}{precision + recall}$
