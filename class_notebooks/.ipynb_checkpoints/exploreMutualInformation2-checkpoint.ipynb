```r
library(infotheo)
```

```r
data = Cars93[, !names(Cars93) %in% c( "Model", "Make", "Max.Price", "Min.Price")]
```

```r
data = na.omit(data)
```

```r
discrete_data = discretize(data, disc = "equalfreq", nbins = 10)
```

```r
target = discrete_data$Price
```

```r
predictor_vars = setdiff(names(discrete_data), "Price")
```

```r
mi_values = sapply(predictor_vars, function(var) {
  mutinformation(discrete_data[[var]], target)
})
```

```r
mi_values_sorted = sort(mi_values, decreasing = TRUE)
```

```r
print(mi_values_sorted)
```

```r
library(minerva) #needs all numeric inputs
```

```r
data = Cars93[, !names(Cars93) %in% c( "Model", "Make", "Max.Price", "Min.Price")]
```

```r
data = na.omit(data)
```

```r
categorical_vars = names(data)[sapply(data, is.factor)]
numerical_vars = names(data)[sapply(data, is.numeric)]
```

```r
data_numeric = data
data_numeric[categorical_vars] = lapply(data_numeric[categorical_vars], function(x) as.numeric(x))

str(data_numeric) # all numeric?..
```

```r
data_onehot = model.matrix(~ . -1, data = data)
```

```r
data_onehot = as.data.frame(data_onehot)
```

```r
target_variable = "Price"

predictor_vars = setdiff(names(data_numeric), target_variable)
```

```r
mic_values = numeric(length(predictor_vars))
names(mic_values) = predictor_vars

for (var in predictor_vars) {
  predictor = data_numeric[[var]]
  target = data_numeric[[target_variable]]
  
  df = data.frame(predictor, target)
  
  mine_result = mine(df)
  
  mic_values[var] = mine_result$MIC[1, 2]
}
```

```r
mic_values_sorted = sort(mic_values, decreasing = TRUE)

print(mic_values_sorted)
```
