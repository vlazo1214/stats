# Plush keychain weights 

Miniso sells Cinnamoroll plush keychains. Today there were 20 plush keychains left in stock. The average weight of them was 45 grams. What they are supposed to weigh when they are produced and leave the factory is 50 grams. We want to figure out if the average weight that we took is significantly different than the weight that they are supposed to weigh, or if it is possible that our average weight is normal.

## Init variables

```r
n_samples = 20
x_bar = 45
mu_0 = 50
sd_pop = 2
alpha = 0.05
num_simulations = 1000
```

## Set hypotheses

H_0: the average weight we took is NOT significantly different from the weight that it is supposed to be (`mu_0 == x_bar`) 

H_A: the average weight we took IS significantly different from the weight that it is supposed to be (`mu_0 != x_bar`)

## Find t stat

```r
t_stat = (mu_0 - x_bar) / (sd_pop / sqrt(n_samples))
t_stat
```
