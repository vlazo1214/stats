## Part 2b: Make a QQ plot comparing the two age samples. 

```r
males = medical_df$age[medical_df$sex == "Male"]
females = medical_df$age[medical_df$sex == "Female"]

qqplot(males, females,
       xlab = "Male ages quantiles",
       ylab = "Female ages quantiles",
       main = "QQ Plot: Male vs Female Age")
abline(0, 1, col = "red")  
```

## Part 2c: Test normality of age within each sex group. 

```r
# get unique values for test
u_males = unique(males)
u_females = unique(females)

# Perform KS test for normality on both samples
ks.test(u_males, "pnorm")
ks.test(u_females, "pnorm")
```

### Given a p-value of 2.2e-16, we reject a null hypothesis of the two samples not following a normal distribution. The two means of male and female ages both follow a normal distribution.

## Part 2d: Test whether the age distributions are the same across sex using a two-sample test (Welch t-test if normality holds, otherwise Wilcoxon rank-sum). 

```r
t.test(males, females)
```

### With a p-value of 0.492, we fail to reject a null hypothesis of the two means being equal to each other. The mean of male and female ages are not significantly different from each other.

## Part 3a: Compute Pearson’s correlation between age and income and test significance. 

```r
# get age and income rows
age_rows = medical_df$age
income_rows = medical_df$income
```

```r
# perform correlation test
cor.test(age_rows, income_rows)
```

### With a p-value of 0.0887, we fail to reject a null hypothesis of the two samples being correlated. Therefore, age and income do not have an association with each other.

## Part 3b: Compute Spearman’s rho and test significance.

```r
cor.test(age_rows, income_rows, method = "spearman")
```

### We started with a null hypothesis of there being neither a significant positive or negative relationship between age and income. With a p-value of 0.09347, we fail to reject this null hypothesis. We can still deduce that there is a positive relationship based on a rho of 0.00530441. This signifies that while the relationship is very weak, it still shows that age is positively related to income.

## Part 6: Fit the model, report coefficients, standard errors, p-values, and R2/Adjusted R2. Comment on significant predictors and interpretation of signs and magnitudes

## Part 8: Fit the logistic model (0/1 response). Report odds ratios with 95% confidence intervals, model deviance, and a simple classification summary at threshold 0.5. Briefly interpret the direction and strength of each predictor. Produce a table to discuss the FP and FN on the table along with the accuracy.

The model I will be creating is using is_high_risk as a response and had_major_procedure, cancer_history and income as the predictors.

```r
ihr_model = glm(is_high_risk ~ had_major_procedure + cancer_history + income, data = medical_df, family = binomial)
summary(ihr_model)
```

```r
# Odds Ratio
OR = exp(coef(ihr_model))
CI = exp(confint(ihr_model))

OR_table = cbind(Odds_Ratio = OR, CI)
round(OR_table, 3)
```

To interpret the efficiency of the predictors, had_major_procedure and cancer_history had a p-value of 2.2e-16 while income had 0.147. We can reject a null hypothesis of there being no significant predictors and conclude there was at least one (had_major_prcedure and cancer_history) that showed a linear relationship with is_high_risk.

To interpret odds ratio, the predictor had_major_procedure shows someone is 1.6 times more likely to be at high risk while cancer_history shows someone is 3.7 times more likely to be at high risk.

To interpret deviance, the residual deviance was lower than the null deviance, indicating the predictors had an improvement on the model.

```r
# predicted probabilities
predicted_prob = predict(ihr_model, type = "response")
predicted_class = ifelse(predicted_prob >= 0.5, "1", "0")
predicted_class = as.factor(predicted_class)

# create confusion matrix
conf_matrix = table(Predicted = predicted_class, Actual = medical_df$is_high_risk)
conf_matrix
```

```r
accuracy = sum(diag(conf_matrix)) / sum(conf_matrix)

FP = conf_matrix["1","0"]
FN = conf_matrix["0","1"]  

list(accuracy = accuracy, false_positives = FP, false_negatives = FN)
```

As shown above, the accuracy of the model is 0.64, produced 687 false positives and 35317 false negatives. This means that the amount of people that were predicted to be high risk but are actually not was 687 and the amount of people that were not predicted to be high risk but actually were was 35317. Looking at model performance as a whole, though, it was able to be 64% accurate.

## Part 2a: Fit a logistic regression to predict basement (yes or no) using four predictors. Justify your choice of predictors in one or two sentences. Report model fit statistics and interpret the direction of effects.

The predictors for basement I am choosing are price, area, stories, bedrooms. Price and stories are to account for an extra house feature, area to account for the extra area of the house, and bedrooms for if there is a family for them being able to store everything for them.

```r
# Standardize capitalization and spaces
num_bsmnt = trimws(tolower(housing_df$basement))

# Convert to numeric 0/1
num_bsmnt = ifelse(num_bsmnt == 'yes', 1, 0)

# create model
bsmnt_model = glm(num_bsmnt ~ price + area + stories + bedrooms, data = housing_df)
summary(bsmnt_model)
```

The predictors that performed from best to worst in regards to standard error were price, area, bedroom, and stories. 

The residual deviance was less than null deviance which shows that the model was improved based on the predictors.

All of the predictors used in this model proved to have a significant effect on the response because all of their p-values were less than the alpha of 0.05. In other words, price, area, stories, and bedrooms all contributed in predicting whether a house has a basement or not.

## Part 2b: Produce a ROC curve, the confusion matrix counts (TP, FP, TN, FN) and accuracy at threshold 0.5, then repeat at threshold 0.25. Describe what changes and explain why a lower threshold might be better or worse in this context

```r
# install.packages("pROC")
library(pROC)
```

```r
# threshold of 0.5
theta = 0.5
p_hat = predict(bsmnt_model)
y_hat = ifelse(p_hat >= theta, 1, 0)
table(Actual = num_bsmnt, Predicted = y_hat)

roc_obj = roc(response=num_bsmnt, predictor=p_hat, quiet=TRUE)
plot(roc_obj, main = "ROC curve")
```

```r
# threshold of 0.25
theta = 0.25
p_hat = predict(bsmnt_model)
y_hat = ifelse(p_hat >= theta, 1, 0)
table(Actual = num_bsmnt, Predicted = y_hat)
```

When a threshold of 0.5 was used, there were 118 false negatives and 29 false positives. When a threshold of 0.25 was used, there were 21 false negatives and 229 false positives. 

In terms of housing data, if someone were to go to an open house expecting there to be a basement and it turns out there is not one present (false positive), that person would be let down. On ther other hand, if they went in not expecting there to be a basement and it turns out there is one (false negative), the property would appear to be more appealing since it would be a pleasant surprise to the person. In this context, false negatives would be better, meaning a higher theta is desirable.

## Part 3a: Create a binary parking present variable that is 1 if parking is greater than 0 and 0 otherwise. Fit a logistic regression for parking present using reasonable structural predictors.

```r
# binary representation of parking
bin_parking = ifelse(housing_df$parking > 0, 1, 0)
```

Predictors: Price, bedrooms, mainroad, and guestroom

```r
# numeric representations
num_mnrd = ifelse(housing_df$mainroad == 'yes', 1, 0)
num_guestroom = ifelse(housing_df$guestroom == 'yes', 1, 0)

# create model
parking_model = glm(bin_parking ~ price + bedrooms + num_mnrd + num_guestroom, data = housing_df)
summary(parking_model)
```

The residual deviance was less than null deviance which shows that the model was improved based on the predictors.

One predictor used in this model proved to have a significant effect on the response because its p-value was less than the  alpha of 0.05. In other words, price was the predictor that contributed in predicting whether a house has parking or not.

## Part 3b: Produce a ROC curve, and report the AUC.

```r
p_hat = predict(parking_model)

roc_obj = roc(response=num_bsmnt, predictor=p_hat, quiet=TRUE)
plot(roc_obj, main = "ROC curve")
auc(roc_obj)
```

An AUC of 0.60 tells us that the model performed slightly better than random 50/50 guessing. It can be concluded that from this model it is hard to tell whether a house has parking or not.

## Part 4a: Build a logistic regression model for unfurnished versus other (anything not ’unfurnished’ becomes other). Fit and compare at least three models. 

Model A predictors: area + stories + prefarea <br>
Model B predictors: area + stories + prefarea + hotwaterheating + num_bsmnt <br>
Model C predictors: area * stories + prefarea + hotwaterheating  + num_bsmnt

```r
# binary rep whether a house is furnished (1) or unfurnished (0)
bin_furnished = ifelse(housing_df$furnishingstatus != 'unfurnished', 1, 0)

# define models A B and C
model_A = glm(bin_furnished ~ area + stories + prefarea, data = housing_df)
model_B = glm(bin_furnished ~ area + stories + prefarea + hotwaterheating + num_bsmnt, data = housing_df)
model_C = glm(bin_furnished ~ area * stories + prefarea + hotwaterheating + num_bsmnt, data = housing_df)
```

```r
# model A summary:
summary(model_A)
```

```r
# model B summary:
summary(model_B)
```

```r
# model C summary:
summary(model_C)
```

Model A only had one predictor that was significant in showing a linear relationship with bin_furnished; models B and C, however, had three and four, respectively. Both had area, stories, and num_bsmnt as significant predictors while model C had an additional: the interaction between area and stories. It stands to note that all three models would reject a null hypothesis of not having a predictor that showed a linear relationship with bin_furnished. Model C is shown to perform the best because it has the most significant predictors of the three models.

## Part 4b: For each model, report AIC and AUC and discuss which model is better and why the included variables make sense. 

```r
p_hat_A = predict(model_A)
roc_obj_A = roc(response=bin_furnished, predictor=p_hat_A, quiet=TRUE)
auc(roc_obj_A)
AIC(model_A)
```

```r
p_hat_B = predict(model_B)
roc_obj_B = roc(response=bin_furnished, predictor=p_hat_B, quiet=TRUE)
auc(roc_obj_B)
AIC(model_B)
```

```r
p_hat_C = predict(model_C)
roc_obj_C = roc(response=bin_furnished, predictor=p_hat_C, quiet=TRUE)
auc(roc_obj_C)
AIC(model_C)
```

When comparing AIC's, model C has the lowest value, making it the best model of the three. When comparing AUC's model B had the highest value, making it the best model for that metric. <br>

Out of all of the metrics used up to this point (p-value AIC, AUC), model C has performed the best of the three models. Although model B has it beat in terms of AUC by 0.001, model C outperforms model B in the other metrics most likely because of the predictors used. While both models had the same predictors, model C included interaction between area:stories, which was the only factor that was different between the two models.
