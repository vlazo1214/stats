# Q1: Crocodile Analysis

## Part 1: Load data

```r
croc_df = read.csv('crocodile_dataset.csv')
```

## Part 2a: Create boxplot for Observed Lengths of River and Swamp

```r
river_rows <- croc_df[croc_df$Habitat.Type == "Rivers", ] # Filter rows with 'Rivers' Habitat Type
swamp_rows <- croc_df[croc_df$Habitat.Type == "Swamps", ] # Filter rows with 'Swamp' Habitat Type
```

```r
river_lengths = river_rows$'Observed.Length' # Get river lengths
swamp_lengths = swamp_rows$'Observed.Length' # Get swamp lengths
```

```r
# Make boxplot of river and swamp lengths
boxplot(river_lengths, swamp_lengths, names = c("Rivers", "Swamps"), ylab = "Lengths (m)")
```

## Part 2b: Statistical test comparing average Observed Lengths of Rivers and Swamps

### Null Hypothesis (H_0): There is no significant difference between the two averages of the Observed lengths between river lengths and swamp lengths <br>
### Alternative Hypothesis (H_A): One average is significantly different from the other 

```r
# We can calculate:
t.test(river_lengths, swamp_lengths)
```

#### With a p-value of 0.03491, we can reject the null hypothesis. In other words, we can say that the average length of crocodiles that come from rivers is significantly different from the average length of crocodiles that come from swamps.

## Part 2c: Statistical test comparing median Observed Lengths of Rivers and Swamps

### Null Hypothesis (H_0): There is no significant difference between the two medians of the Observed lengths between river lengths and swamp lengths <br>
### Alternative Hypothesis (H_A): One median is significantly different from the other 

```r
# We can calculate:
wilcox.test(river_lengths, swamp_lengths)
```

#### With a p-value of 0.08229, we fail to reject the null hypothesis. This ultimately means that the median observed length for a crocodile that would come from a river is not significantly different than one that would come from a swamp. The distributions do not differ between observed lengths.

## Part 3a: Create a boxplot of the Observed lengths of crocodiles that were sampled from India and Mexico

```r
india_rows <- croc_df[croc_df$Country.Region == "India", ] # Filter rows with 'Rivers' Habitat Type
mexico_rows <- croc_df[croc_df$Country.Region == "Mexico", ] # Filter rows with 'Swamp' Habitat Type
```

```r
india_lengths = india_rows$'Observed.Length' # Get lengths sampled from India
mexico_lengths = mexico_rows$'Observed.Length' # Get lengths sampled from Mexico
```

```r
boxplot(india_lengths, mexico_lengths, names = c("India", "Mexico"), ylab = "Lengths (m)")
```

## Part 3b: Statistical test comparing average Observed Lengths sampled from India and Mexico

### Null Hypothesis (H_0): There is no significant difference between the two averages of the Observed lengths between lengths sampled from India and lengths sampled from Mexico <br>
### Alternative Hypothesis (H_A): One average is significantly different from the other 

```r
# We can calculate:
t.test(india_lengths, mexico_lengths)
```

#### With a p-value of 8.303e-05, we can reject the null hypothesis. In other words, we can say that the average length sampled from India is significantly different from the average length of sampled from Mexico.

## Part 3c: Statistical test comparing median Observed Lengths sampled from India and Mexico

### Null Hypothesis (H_0): There is no significant difference between the two medians of the Observed lengths between lengths sampled from India and lengths sampled from Mexico <br>
### Alternative Hypothesis (H_A): One median is significantly different from the other 

```r
# We can calculate:
wilcox.test(india_lengths, mexico_lengths, exact=FALSE)
```

#### With a p-value of 8.13e-05, we reject the null hypothesis. This ultimately means that the median observed length for a crocodile that would come from a river is significantly different than one that would come from a swamp. The distributions greatly differ between both observed lengths.

## Part 4a: Produce new column with categorical "High/Low" <br> 
## Part 4b: Produce table with "High/Low" counts and River + Swamp habitat types


```r
# Produce new High/Low column based on mean Observed Lengths
obs_lengths_mean = mean(croc_df$Observed.Length..m)

croc_df$high_low = factor(ifelse(croc_df$Observed.Length..m >= obs_lengths_mean, "High", "Low"))
```

```r
# Get subset
river_swamp_sub_data <- subset(croc_df, Habitat.Type %in% c("Rivers", "Swamps"))

# Produce Table 
rs_hl_table = table(river_swamp_sub_data$high_low, river_swamp_sub_data$Habitat.Type)
rs_hl_table
```

## Part 4c: Produce a Chi-square test and find CramerV

### Null Hypothesis (H_0): The River and Swamp Habitat types are not associated with High and low lengths <br>
### Alternative Hypothesis (H_A): There is a clear association between these two categories 

```r
# Perform test
rs_hl_test = chisq.test(rs_hl_table)
rs_hl_test
```

#### With a p-value of 0.6358, we fail to reject the null hypothesis. This ultimately means that there is no clear association between river and swamp habitat types and high and low lengths.

```r
if (!require("vcd")) install.packages("vcd")
library(vcd)

# Cramer V
cramer_v = assocstats(rs_hl_table)$cramer
cramer_v
```

### This CramerV value is close to 0, meaning there is weak association.

### Part 4d: Produce a 2 way table with the High/Low for the Habitat Type of Rivers and Swamps against Conservation Status Vulnerable/Critically Endangered

```r
# Get subset
hl_rs_vc_subdata = subset(croc_df, Conservation.Status %in% c("Vulnerable", "Critically Endangered") & Habitat.Type %in% c("Rivers", "Swamps"))

hl_rs_vc_table = table(hl_rs_vc_subdata$high_low, hl_rs_vc_subdata$Conservation.Status)
hl_rs_vc_table
```

### Part 4e: Perform Chi square test with and without Yates

### Null Hypothesis (H_0): The High/lows of River and Swamp Habitat types are not associated with Conservation statuses Critically endangered/Vulnerable <br>
### Alternative Hypothesis (H_A): There is a clear association between these two categories 

```r
# Perform test with Yates correction:
chisq.test(hl_rs_vc_table)
```

```r
# Perform test without Yates correction:
chisq.test(hl_rs_vc_table, correct=FALSE)
```

#### With both p-values being less than 0.05 with and without Yate's correction, we reject the null hypothesis. This ultimately means that there is an association with high/low lengths of rivers and swamps and Conservation statuses Critically endangered/Vulnerable

```r
# Cramer V
cramer_v = assocstats(hl_rs_vc_table)$cramer
cramer_v
```

### This Cramer V indicates a moderate association between these two categories.

# Q2: Indian Water Quality

## Part 1: Load data

```r
water_df = read.csv('Indian_water_data.csv')
```

## Part 2: Test if 'Dissolved - Max' values are normally  distributed

```r
# Get 'Dissolved - Max' rows
dis_max_rows = water_df$Dissolved...Max
```

### Null Hypothesis (H_0): The values of Dissolved - Max do NOT follow a normal distribution <br>
### Alternative Hypothesis (H_A): The values of Dissolved - Max DO follow a normal distribution

```r
shapiro.test(dis_max_rows)
hist(dis_max_rows)
```

#### With a p-value of 1.722e-06, we reject the null hypothesis. This ultimately means that the values of dissolved - max follow a normal distribution

## Part 3: Test if the mean of Dissolved Max rows is 7.0 (Since we found the data to be normally distributed)

### Null Hypothesis (H_0): The mean of Dissolved - Max values is 7.0 <br>
### Alternative Hypothesis (H_A): The mean of Dissolved - Max values is NOT 7.0

```r
t.test(dis_max_rows, mu=7)
```

#### The p-value is 1.099e-06, we reject the null hypothesis. This means we can say the mean is not equal to 7.0. 

## Part 4a: Plot distributions for Dissolved max for Type Water Body's River and Drain

```r
# Get dissolved max rows for river and drain
river_type_rows = water_df[water_df$Type.Water.Body == "RIVER", ]
drain_type_rows = water_df[water_df$Type.Water.Body == "DRAIN", ]

# Histograms for each
hist(river_type_rows$Dissolved...Max, xlab="Dissolved Max levels")
hist(drain_type_rows$Dissolved...Max, xlab="Dissolved Max levels")
```

### Part 4b: Test for Normality for each

```r
# Normality test for River Diss. Max
shapiro.test(river_type_rows$Dissolved...Max)
```

```r
# Normality test for River Diss. Max
shapiro.test(drain_type_rows$Dissolved...Max)
```

### P-values from both tests allow us to reject a null hypothesis of either of them not following a normal distribution. This denotes that dissolved max values from river and drains follow normal distributions.

## Part 4c: T-test comparing if their means are the same (since we found that they follow normal distributions)

```r
t.test(river_type_rows$Dissolved...Max, drain_type_rows$Dissolved...Max)
```

### We started with a null hypothesis of the means of dissolved max values for river and drain being the same. We see that the p-value is large, which means we fail to reject the null hypothesis. In other words, both means are not significantly different from each other.

## Part 5: Find correlation of Temperature Max and pH - Max

```r
# Get rows of each:
tempmax_rows = water_df$Temperature..C....Max
phmax_rows = water_df$pH...Max
```

```r
# Perform test:
cor.test(tempmax_rows, phmax_rows)
```

### Since we started with a null hypothesis of no association between max temperature and ph max, we can reject that null hypothesis because of the p-value we found. This ultimately means that there is an association between max temperature and max ph

## Part 6: Perform regression between Temperature and ph max by using F-test

```r
# Regression model:
temp_ph_model = lm(phmax_rows ~ tempmax_rows, data = water_df)

summary(temp_ph_model)
```

```r
# Produce a plot
plot(temp_ph_model)
```

### Given the p-value of the regression/F-test, we can conclude that we reject the null hypothesis of temperature and ph max values not having an association with each other. This shows that these two sets of values have a significant relationship with each other.

# Q3: Diabetes data

## Part 1: Load the data

```r
diabetes_df = read.csv('diabetes.csv')
```

## Part 2: Filter out impossible values

```r
# Effectively keeps rows that dont have impossible data 
clean_diabetes_df = subset(diabetes_df, BloodPressure != 0 & Insulin != 0 & SkinThickness != 0 & BMI != 0)
```

## Part 3: Find columns that have a normal distribution

```r
# List to keep track of whether a column has a normal distribution
normal_dists = list()

# Iterate over each column and appending to a list which columns are normally distributed
for (col in colnames(clean_diabetes_df)) {
    curr_pval = shapiro.test(clean_diabetes_df[[col]])$p.value
    normal_dists[[col]] = curr_pval
}

print(normal_dists)
```

### All of the columns in the dataframe follow a normal distribution. 

## Part 4: Histograms for Glucose, Blood pressure, and Outcome

```r
# Histograms for the base columns
hist(clean_diabetes_df$Glucose)
hist(clean_diabetes_df$BloodPressure)
```

```r
# Create subsets only looking at Outcome 1 and Outcome 0
outcome_1_sub = subset(clean_diabetes_df, Outcome == 1)
outcome_0_sub = subset(clean_diabetes_df, Outcome == 0)

hist(outcome_1_sub$Glucose, 
     col = rgb(0, 0, 1, 0.4), 
     xlim = c(0, 200), 
     freq = FALSE,
     main = "Glucose vs Blood Pressure",
     xlab = "Value")

hist(outcome_1_sub$BloodPressure, 
     col = rgb(1, 0, 0, 0.4), 
     freq = FALSE,
     add = TRUE)
```

hist(outcome_0_sub$Glucose, 
     col = rgb(0, 0, 1, 0.4), 
     xlim = c(0, 200), 
     freq = FALSE,
     main = "Glucose vs Blood Pressure",
     xlab = "Value")

hist(outcome_0_sub$BloodPressure, 
     col = rgb(1, 0, 0, 0.4), 
     freq = FALSE,
     add = TRUE)

### The second histogram shows the two distributions are more significantly similar while the first shows the two distributions are more significantly different. In the context of the data, Blood pressure and Glucose for outcome 1 are shown to be distributed more far apart than outcome 0.

## Part 5: Blood pressure quantiles

```r
bp_quants = quantile(clean_diabetes_df$BloodPressure, probs = c(0.25, 0.5, 0.75))
```

```r
clean_diabetes_df$bp_factor <- cut(
  clean_diabetes_df$BloodPressure,
  breaks = c(-Inf, bp_quants[1], bp_quants[2], Inf),
  labels = c("Low", "Medium", "High"),
  right = TRUE
)
```

```r
# Produce ANOVA
anova_model <- aov(Glucose ~ bp_factor, data = clean_diabetes_df)

summary(anova_model)
```

### With a p-value < 0.05, we reject the null of Glucose values being consistent with Blood pressure category. There is at least one of 'Low', 'Medium', and 'High' that has a different mean Glucose
